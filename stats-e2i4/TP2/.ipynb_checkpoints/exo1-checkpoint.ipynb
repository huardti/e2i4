## Statistiques descriptives 2

les données sont dans le fichier Carsmall3.xlsx

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# On charge les données
X = pd.read_excel("Carsmall3.xlsx", sheet_name=0, header=0, index_col=0, engine="openpyxl")
```

# taille du jeu de données

le code suivant donne le nombre de voiture dans le jeu de données ainsi que le nombre de variable presente pour chaque voiture

```python
nombre_voiture, nombre_var = X.shape
print("nombre de voiture:", nombre_voiture, "\nnombre de variables:", nombre_var)
```

# Variables discrètes / quantitatives

# Description de chaque variable

Modèle:

Origine

```python
plt.figure()
(Valeurs,FreqAbsolues)=np.unique(X["Origin"], return_counts=True)
plt.bar(Valeurs,X["Origin"])
plt.show()
```

Poid:

```python
plt.figure()
plt.boxplot(X['Weight'])
plt.xlabel("poid")
plt.show()
```

```python
print("moyenne", np.mean(X["Weight"]))
print("Ecart type", np.std(X["Weight"]))

FreqRelatives, bin_edges = np.histogram(X["Weight"], bins=20, density=True)

plt.figure()
plt.bar(bin_edges[:-1], FreqRelatives, width=bin_edges[1]-bin_edges[0], color='red', alpha=0.5)
plt.xlabel('Poid')
plt.ylabel('Effectifs relatifs')
plt.show()
```

# Lien entre Weight et MPG

on attribu les couleurs en fonction de l'origine

```python
Origin_tab = X["Origin"].str.strip()
origin_color = ["red"if o == "USA" else "blue" if o == "Europe" else "green" for o in Origin_tab]
```

```python
X.plot.scatter(x='Weight',y='MPG', c=origin_color)
```
